```python
#1. Tokenize
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")

sequence = "Here is the example of tokenizer"
tokens = tokenizer.tokenize(sequence)

print(tokens)
```

```python
#2. As input ids
input_ids = tokenizer.convert_tokens_to_ids(tokens)
```

```python
#Also decode by reverse-engineering
decoded_string = tokenizer.decode(input_ids)
print(decoded_string)
```

#2. Multiple Sequences

```python
import tensorflow as tf
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification

checkpoint = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = TFAutoModelForSequenceClassification.from_pretrained(checkpoint)

sequence = "I've been waiting for a HuggingFace course my whole life."

tokens = tokenizer.tokenize(sequence)
ids = tokenizer.convert_tokens_to_ids(tokens)

#input_ids = tf.constant(ids) // model(input_ids)
input_ids = tf.constant([ids]) #added another dimension
print("Input IDs:", input_ids)

output = model(input_ids)
print("Logits:", output.logits)
```

Which models have the output logits?

At least for **DistilBERT**, that includes the input IDs as well as the attention mask.

# Having it altogether

```python
import tensorflow as tf
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = TFAutoModelForSequenceClassification.from_pretrained(checkpoint)
sequences = ["I've been waiting for a HuggingFace course my whole life.", "So have I!"]

#padding,trunctation to make this long
tokens = tokenizer(sequences, padding=True, truncation=True, return_tensors="tf")
output = model(**tokens)
```
