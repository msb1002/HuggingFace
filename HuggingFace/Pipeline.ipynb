```python
#Simply run the pipeline after specifying which pipeline to use
from transformers import pipeline
classifier = pipeline("sentiment-analysis")
e1 = classifier("Hello my name is Sun Bin, and I am very excited to start this code!")
e1
```

```python
#There can be multiple
e2 = classifier(["hello","I absolutely hate it"])
e2
```

```python
classifier_2 = pipeline("zero-shot-classification")
e_3 = classifier_2("More election posts coming up soon",candidate_labels = ["entertainment","news","politics"])
e_3
```

```python
#Also specify the type of model usong the model= part
from transformers import pipeline

generator_1 = pipeline("text-generation", model="distilgpt2")
generator_1(
    "How is this even possible?",
    max_length=30,
    num_return_sequences=2,
)
```

```python
from transformers import pipeline

unmasker = pipeline("fill-mask")
unmasker("This course will <mask> you all about mathematics.", top_k=4)
```

```python
#NER - parse into persons, locations, or organizations
from transformers import pipeline

question_answerer = pipeline("question-answering")
question_answerer(
    question="Who can speak two languages?",
    context="Harry can speak both Chinese and English. However, Sun can only speak Korean",
)
```
